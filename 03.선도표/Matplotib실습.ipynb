# 3. 선도표(line plot)그리기

### 3.1 기본적인 그래프 그리기

```
plt.plot([x], y, [fmt], [x2], y2, [fmt2] ..., **kwargs)
fmt: 색상, 마커, 선 스타일과 같은 기본 서식 설정
(하나의 plotdp 여러 개의 좌표와 fmt를 설정하면 복수개의 그래프를 그릴 수 있다.)
```
- plot() 함수의 인자로 표시할 데이터를 전달한 후 `show()` 함수를 통해 출력한다.
- plot() 함수에 전달되는 리스트는 자동으로 y축의 값으로 설정된다.
- x축의 값은 0으로 시작해서 1의 간격으로 자동으로 설정된다.

````python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'D2Coding ligature'


plt.plot([0,2,4,5],'-.')
plt.show()
````

- 두 개의 리스트를 전달함으로써 각각 x축의 값, y축의 값으로 설정된 그래프를 그릴 수도 있다.

````python
plt.plot([2,4,6,8],[1,16,4,9])
# 축의 범위 [xmin,xmax,ymin,ymax]
plt.axis([0,10,0,20])
plt.show()
````

# 스타일
- plot() 함수에 다양한 키워드 인수를 전달하여 스타일을 바꿀 수 있다.
  - linestyle: solid, dashed, dotted, dash-dot

````python
import numpy as np
plt.plot(np.arange(6), #x
         np.arange(6)**2, # y
         color='red', # 스타일
         linestyle='dashed',
         linewidth=2.5,
         marker='o',
         markeredgecolor='green',
         markerfacecolor='yellow',
         markersize=9.3
         )
plt.show()
````

- 상징(각 스타일을 나타내는 단축 문자열)을 이용해서 문자열 형태로 설정할 수도 있다.

````python
# linspace(초기값0, 까지,10등분 )
x = np.linspace(0, np.pi*2, 10)
# r: 레드 o:점  --:실선                     ^: 세모 -.:
plt.plot(x, np.sin(x), 'r--o', x, np.cos(x),'g^-.',x , -np.sin(x),'bx:')
plt.xlim(0, np.pi*2)
plt.show()
````

- 각표의 상징은 다음과 같다.


````python
plt.plot([1,2,3,4],[5,20,24,56],'bx--')
plt.title('Line Plot')
plt.xlabel('X data')
plt.ylabel('Y data')
plt.show()
````

````python
data = np.arange(0, 5,0.2)

plt.plot(data, data, 'r--', data, data**2, 'bs', data,data**3,'g^-.')

plt.ylim(0, 100)
plt.xlim(0, 5)
plt.title('Line Plot')
plt.xlabel('X data')
plt.ylabel('Y data')
plt.show()
````

````python
x= np.arange(10)
y= np.random.randint(0,100,10)

plt.plot(x, y, c = 'b',lw=3,ls=':',marker='o', ms =10, mec='r', mew=2, mfc='y')
plt.title('Various Style')
plt.show()
````

# 범례 표시하기

- plot() 함수에 label 인수를 통해 그래프에 이름을 정할 수 있다.
- 이렇게 지정된 이름은 plt.legend() 함수를 통해 범례로써 표시할 수 있다 .
- loc 파라미터에 튜플로 두 개의 값을 지정하면 해당 위치에 범례가 표시된다(0~1사이의 값)
```
'best' (default값)
'upper right'
'upper left'
'lower right'
'lower left'
'right'
'center left'
'center right'
'lower center'
'upper center'
'center'
```

````python
a = np.arange(0,3,0.02)

#지수함수를 이용해 변환된값
c=np.exp(a) 

d = c[::-1]
plt.plot(a,c,'r--',label='Model length')
plt.plot(a,d,'b:',label="Data length")
plt.plot(a,c+d,'g',label="Total message length")
# 범례가 표시될 위치 loc파라미터 이용,shadow=범례에 그림자 데코레이터
plt.legend(loc='best') 
plt.show()
````

# 서브플롯 그리기

- plt.subplot() 함수를 이용해서 하나의 figure에 여러개의 그래프를 그릴 수 있다.
- 주요인자
    - nrow, ncols, index를 지정하는 3가지 정수를 이용하여 서브플롯의 위치를 설정한다.

````python
x1 = np.linspace(0, 5)
x2 = np.linspace(0, 2)

y1 = np.cos(2*np.pi * x1) * np.exp(-x1)
y2 = np.cos(2*np.pi * x2)

plt.figure(figsize=(11,4))
plt.subplot(2,1,1) # nrows=2, ncols=1, index=1
plt.plot(x1,y1,'o-')
plt.title('1st Graph')
plt.xlabel('tims (s)')
plt.ylabel('Damped')
plt.subplot(2,1,2) # nrows=2, ncols=1, index=2
plt.plot(x2,y2,'.-')
plt.title('2st Graph')
plt.xlabel('tims (s)')
plt.ylabel('Undamped')
# 서브플롯의 레이아웃을 자동으로 맞춰준다.
plt.tight_layout() 
plt.show()
````

- plt.subplots() 함수를 이용해서 서브플롯을 그릴 수 있다.
- subplots() 함수는 Figure와 Axes 객체를 반환한다.
    - Axes 클래스: 공간이 확보된 이미지 영역 정보를 가지고 있다.
- 주요 인자
    - nlows, ncols: 정수 값으로 서브플롯의 행,열수 지정
    - sharex, sharey: boolean값(default: False) x,y축 사이에 공유 설정

````python
fig, ax = plt.subplots(1, 4, sharey=True)
#figure 사이즈 설정
fig.set_size_inches((16,4)) 

# 산점도 ?
ax[0].scatter(np.random.rand(100),np.random.rand(100),c = 'cyan', edgecolor='red',alpha=0.5) # scatter=산점도
ax[1].scatter(np.random.rand(100),np.random.rand(100),c = 'magenta', edgecolor='green',alpha=0.5)
ax[2].scatter(np.random.rand(100),np.random.rand(100),c = 'yellow', edgecolor='blue',alpha=0.5)
ax[3].scatter(np.random.rand(100),np.random.rand(100),c = 'white', edgecolor='black',alpha=0.5)

ax[0].set_title('subplot 0')
ax[1].set_title('subplot 1')
ax[2].set_title('subplot 2')
ax[3].set_title('subplot 3')

plt.show()
````

# 막대 그래프(bar chart) 그리기
- plt.bar()함수를 이용해서 막대 그래프를 그릴 수 있다.

````python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0,20)
y = x*2


#라벨은 나중에 범례에도 쓰기위함,linewidth는 테두리 
plt.bar(x,y, label='Blue Bar',color='b',edgecolor='red',linewidth=.5) 

plt.xlabel('bar number')
plt.ylabel('bar height')
plt.title('Bar Chart Example')

#(축 인덱스, 새로운 축 번호 값)
xticks=(np.arange(len(x)))

plt.xticks(xticks,xticks+1) 
plt.legend()
plt.show()
````

- plt.barh()함수를 이용해서 수평 막대 그래프를 그릴 수 있다.

````python
x = np.arange(0,20)
y = x*2

#라벨은 나중에 범례에도 쓰기위함,linewidth는 테두리 
plt.barh(x,y, label='Blue Bar',color='b',edgecolor='yellow',linewidth=3) 
plt.ylabel('bar number')
plt.xlabel('bar height')
plt.title('Bar Chart Example')
plt.yticks(np.arange(len(x)),np.arange(len(x))+1)
plt.legend()
plt.show()
````

- matplotlib 색상표


````python
labels = ['Seoul','Busan','Incheon','Gwangju','Jeju']
men_means   = [20,34,30,35,27]
women_means = [25,32,34,20,25]
x = np.arange(len(men_means)) # [0,1,2,3,4]
width = 0.2

plt.bar(x-width/2, men_means, width, color='dodgerblue', label='Men')
plt.bar(x+width/2, women_means, width, color='violet', label='Women')
plt.title('Score by Group and gender')
plt.xlabel('city')
plt.ylabel('score')
plt.xticks(x,labels)
plt.legend()
plt.show()
````

# 산점도
- 산점도는 직교 좌표계를 이용해 두 개 변수 간의 상관관계를 나타내는 플롯이다.
- plt.scatter()함수를 이용해서 산점도를 그릴 수 있다.

````python
# 산점도라는건 회귀 분석(regression)을 할 때
# 둘 이상의 변수 간의 관계를 보여주는 통계적 방법
x1 = [2,3,4]
y1 = [5,5,5]

x2 = [1,2,3,4,5]
y2 = [2,3,2,3,4]

y3 = [6,8,7,8,7]

plt.scatter(x1,y1)
plt.scatter(x2,y2, marker='v',color='g')
plt.scatter(x2,y3, marker='^',color='r')
plt.title("Scatter Plot Example")
plt.show()
````

````python
np.random.seed(0)

for color in ['blue','orange','green']:
    n = 50
    x, y = np.random.rand(2,n)
# 10이상 200미만의 난수 50개 생성 == [10,200)  (scale:동그라미)
    scale = 190.0 * np.random.rand(n) + 10
    plt.scatter(x,y,c=color,s=scale,label=color,alpha=.3,edgecolors='none')
plt.legend()
# grid = 격자 생성
plt.grid(True)
plt.show()
````

# 히스토그램 그리기
- plt.hist() 함수를 이용해서 히스토그램을 그릴 수 있다. (도수분포표)
- 자료의 분포를 몇개의 구간으로 나눈 후 각 구간에 속하는 데이터의 빈도수를 표로 나타낸 것을 도수분포표라고 한다.(frequency table)
- 히스토그램은 도수 분포표에서 구간별 빈도수를 막대 그래프로 나타낸 것이다.


````python
y = np.random.randn(1000) # 평균 0, 표준편차가 1인 표준 정규분포로부터 난수 1000개를 생성

x = np.arange(len(y))
plt.bar(x, y)
plt.title('Raw Data<Bar Chart>')
plt.show()

# 표현할 데이터를 50개의 구간으로 나눈다.
plt.hist(y,bins=50) 
plt.title('Histogram')
plt.show()
# 누계 커멀러티브
plt.hist(y, bins=10,cumulative=True)
plt.title('Cumulative Histogram')
plt.show()
````

# 원 그래프(pie chart) 그리기
- plt.pie()함수를 이용해서 원그래프를 그릴 수 있다.
- pie() 함수에 전달하는 값은 굳이 비율 값을 전달하지 않고 일반적인 숫자 값을 전달하더라도 값들의 총 합에 대한 비율을 자동으로 계산한다.

````python
x =10.22
print('%.1f'%x)
print('%.1f%%'%x) #출력에 %를 붙히고 싶을때
````

````python
labels=['Python','Java','C']
# autopct : wedge(파이의 조각)위에 표시할 레이블 형식(default:None) 자동으로 수치를 100으로 놓고 계산해줌
plt.pie([0.3,0.35,0.4], labels = labels, autopct='%.1f%%')
plt.title('Popular Language Ranking')
plt.show()
````

````python
sections= [0.3,0.35,0.4]
labels=['Python','Java','C']
color = ['coral','dodgerblue','yellowgreen']
explode = [0,.1,0]

plt.pie(
        # 조각의 크기를 설정하는 값
        sections,
        # 조각을 중심점으로부터 얼마만큼 떨어지게 할 것인지 설정 (반지름의 10%만큼 벗어나남을 의미)
        explode,
        # 조각의 이름을 표시
        labels= labels,
        # 비율을 표시하는 형식 문자열
        autopct='%.1f%%',
        #조각의 색깔을 결정
        colors=color,
        # 그림자를 표시
        shadow=True,
        # 원형의 시작부분을 각도로 표시 (기본값은 0, 3시방향)
        startangle=0, 
        counterclock=False, # 반 시계 방향으로 순서대로 표시할지 결정(기본값은 True)
        rotatelabels=False # 조각의 이름을 회전시킨다(default =False)
       )
plt.show()
````

# 박스 플롯 그리기
- plt.boxpolt(): 많은 데이터에서 데이터의 범위와 중앙값 등 통계 수치 값을 확인하고자 하는 목적으로 사용한다.
- 이상치(Outlier)가 있는지 검수하기 수월하다.
  - 최소값: 제 1사분위에서 1.5IQR
  - (Inter Quantile Range: 사분위수범위, Q3-Q1)을 뺀 위치
  - 제 1사분위(Q1): 중앙값 기준으로 하위 50%중의 중앙값. 
  - 즉, 전체데이터중 하위 25%에 해당하는 값
  -
